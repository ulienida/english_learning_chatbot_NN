```python
import pandas as pd
```

```python
import textstat
```

```python
prompts_dataset= pd.read_csv('/Users/rakshitha/Downloads/Prompts_Eval.csv')
```

```python
len(prompts_dataset)
prompts_dataset.head(10)
```

```python
prompts_dataset["Prompts"]
```

```python
prompts_df = pd.DataFrame(prompts_dataset)
prompts_df= prompts_df['Prompts']
prompts_df.columns=['Prompts']
prompts_df = prompts_df.to_frame('Prompts')
prompts_df['readability_consensus'] = prompts_df['Prompts'].apply(textstat.text_standard, float_output=False)
prompts_df['readability_score'] = prompts_df['Prompts'].apply(textstat.dale_chall_readability_score)
```

```python
prompts_df.head(5)
```

```python
beginner_dataset_df = prompts_df[prompts_df['readability_score'] <= 6.99]
beginner_dataset_df = beginner_dataset_df['Prompts']
beginner_dataset_df = beginner_dataset_df.to_frame('Prompts')
beginner_dataset_df['readability_score'] = beginner_dataset_df['Prompts'].apply(textstat.dale_chall_readability_score)
beginner_dataset_df.to_csv('begprompts_dataset.csv',encoding='utf-8',index=False)
beginner_dataset_df.head(30)
```

```python
advanced_dataset_df = prompts_df[prompts_df['readability_score'] > 6.99]
advanced_dataset_df= advanced_dataset_df['Prompts']
advanced_dataset_df= advanced_dataset_df.to_frame('Prompts')
advanced_dataset_df['readability_score'] = advanced_dataset_df['Prompts'].apply(textstat.dale_chall_readability_score)
advanced_dataset_df.to_csv('advprompts_dataset.csv',encoding='utf-8',index=False)
advanced_dataset_df.head(30)
```
